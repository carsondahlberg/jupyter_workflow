## get bike data

```python
bike_data_url = 'https://data.seattle.gov/api/views/65db-xm6k/rows.csv?accessType=DOWNLOAD'
```

```python
# to read web data
import io
import requests

# data handling
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# get the data
r = requests.get(bike_data_url).content

# read into df, set dates as the index
df = pd.read_csv(io.StringIO(r.decode('utf-8')), 
                 index_col='Date', 
                 parse_dates=True)

# quick data qual check
display(df.head())
display(df.info())
```

## visualize data

```python
df.resample('W').sum().plot();
```
